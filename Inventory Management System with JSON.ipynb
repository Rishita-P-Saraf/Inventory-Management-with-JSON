```python
import json
import time

#Initializing Variables
sale=''

#Loading the Inventory Records from JSON to string
fd=open("Records.json","r")
js=fd.read() #js is str
fd.close()

#Converting string records to JSON/dictionary
record=json.loads(js) #js becomes dictionary here

#Displaying the MENU
print("----------------MENU-----------------\n")
for key in record.keys():
    print(key, ":", record[key]["Name"], "|", record[key]["Price"], "|",  record[key]["Qn"])
print("-------------------------------------\n")

#Taking user inputs about the purchase and user details
ui_name=str(input("Enter your name     : ")) #since all the inputs are in int
ui_mail=str(input("Enter your Mail ID  : "))
ui_ph  =str(input("Enter your Phone no : "))
ui_pr  =str(input("Enter Product ID    : ")) #since keys are in str in dictonary 
ui_qn  =int(input("Enter Quantity      : "))

#If we have enough quantity
if(record[ui_pr]['Qn']>=ui_qn):

    #Displaying Purchase Details
    print("---------------------------------------")
    print("                 BILL                  ")
    print("Name            : ", record[ui_pr]["Name"])
    print("Price           : ", record[ui_pr]["Price"],"Rs")
    print("Quantity Req    : ", ui_qn)
    print("---------------------------------------")
    print("Billing Amount  : ",record[ui_pr]["Price"]*ui_qn,"Rs")
    print("---------------------------------------")

    #Generating sales
    sale=ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(ui_qn)+","+str(record[ui_pr]["Price"])+","+str(ui_qn*record[ui_pr]["Price"])+","+time.ctime()+"\n"
    
    #Updating the Records dictionary
    record[ui_pr]["Qn"]=record[ui_pr]["Qn"]-ui_qn

#If we dont have enough quantity
else:
    #Displaying info about the available quantity
    print("Sorry, we're not having enough quantity in our inventory")
    print("We have only",record[ui_pr]["Qn"],"quantity")    
    print("---------------------------------------")
    
    ch=str(input("Press Y/y to purchase: ")) #since by default its an int

    #If user wants to purchase whole quantity available of that product
    if (ch=="Y" or ch=="y"):
        #Displaying Purchase Details 
        print("---------------------------------------")
        print("                 BILL                  ")
        print("Name              : ", record[ui_pr]["Name"])
        print("Price             : ", record[ui_pr]["Price"],"Rs")
        print("Quantity Available: ", record[ui_pr]["Qn"])
        print("---------------------------------------")
        print("Billing Amount  : ",record[ui_pr]["Price"]*record[ui_pr]["Qn"],"Rs")
        print("---------------------------------------")

        #Generating sales
        sale=ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(record[ui_pr]["Qn"])+","+str(record[ui_pr]["Price"])+","+str(record[ui_pr]["Qn"]*record[ui_pr]["Price"])+","+time.ctime()+"\n"

        #Updating the Records dictionary
        record[ui_pr]["Qn"]=0

    #If user presses anything other than Y/y
    else:
        print("Thanks!")
        
#Converting dictionary to string so that we can save it to JSON dictionary(dumps func)
js=json.dumps(record)

#json.dump() -to Save to file
#json.loads() -convert string to dictionary
#json.dumps()- convert dictionary to json string

#Saving Records in JSON file
fd=open("Records.json","w")
fd.write(js)
fd.close()

#Saving Records in Sales2 file
fd=open("Sales2.txt","a")  #Change to Sales2.csv when u have to evaluate the sales
fd.write(sale)
fd.close()

print("---------------------------------------------")
print("  Thanks for your order, Inventory updated!  ")
print("---------------------------------------------")
```
